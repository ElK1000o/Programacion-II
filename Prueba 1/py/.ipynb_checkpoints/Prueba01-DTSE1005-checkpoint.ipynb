**7) (15 pts) Encuentra todos los listados de tipo de habitación <font color='red'>Shared room</font> en <font color='red'> Manhattan</font> con un rango de precios de <font color='red'> \\$ 60 a  \\$ 75</font> y que han recibido más de <font color='red'>100 reseñas</font> ("number_of_reviews")**

```python
manhattan = abnb[abnb['neighbourhood_group'] == 'Manhattan']
manhattan = manhattan[manhattan["number_of_reviews"]>100]
manhattan = manhattan[manhattan["price"]>=60]
manhattan = manhattan[manhattan["price"]<=75]
manhattan = manhattan[manhattan["room_type"]=="Shared room"]
manhattan[["room_type", "price", "number_of_reviews"]]
```

**8) (20 pts) Encuentre los valores <font color='red'>mínimo</font>, <font color='red'>máximo</font> y <font color='red'>promedio</font> para <font color ='red'>precio</font> "price" y <font color='red'> número de reseñas</font> ("number of reviews") para los listados en cada uno de los diferentes vecindarios ("neighbourhood") de Manhattan**

Media de precios por vecindario

```python
manhattan1 = abnb[abnb['neighbourhood_group'] == 'Manhattan']
media_p = manhattan1.groupby("neighbourhood")["price"].mean()
min_p = manhattan1.groupby("neighbourhood")["price"].min()
max_p = manhattan1.groupby("neighbourhood")["price"].max()
p_media = pd.DataFrame(media_p)
p_media
```

Minimo de precios por vecindario

```python
p_min = pd.DataFrame(min_p)
p_min
```

Maximo de precios por vecindario

```python
p_max = pd.DataFrame(max_p)
p_max
```

Media de reviews por vecindario

```python
media_n = manhattan1.groupby("neighbourhood")["number_of_reviews"].mean()
min_n = manhattan1.groupby("neighbourhood")["number_of_reviews"].min()
max_n = manhattan1.groupby("neighbourhood")["number_of_reviews"].max()
n_media = pd.DataFrame(media_n)
n_media
```

Minimo de reviews por vecindario

```python
n_min = pd.DataFrame(min_n)
n_min
```
