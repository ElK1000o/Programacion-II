<b>1) (20 pts) Recrea la siguiente figura usando la librería Matplotlib.pyplot y Numpy</b>

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
bx = fig.add_axes([0.43, 0.3, 0.25, 0.25]) 

x = np.linspace(-4, 4, 40)
ay = 3 * np.cos(0.6 * x)
by = 2.8 * np.exp(-0.15 * (x - 0.1)**2)

ax.plot(x, ay, "r-.", label="$3cos(0.6x)$", linewidth=2)
ax.plot(x, by, "b-o", markeredgecolor='black', markerfacecolor='yellow', markersize=8, mew=0.8, label="$2.8e^{-0.15(x-0.1)^2}$")

ax.legend(loc='upper right')
ax.set_title("Gráfica 1 (figsize 5,5)")
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.set_ylim(-1.0, 3.0)
ax.set_xlim(-4, 4)

texts = {'Rojo': (-1.7, 1.7, 'red'), 'Azul': (-0.8, 0, 'blue'), 'Verde': (0.5, -2, 'green')}

for text, (x_coord, y_coord, color) in texts.items():
    bx.text(x_coord, y_coord, text, color=color, fontsize=10)

bx.set_xlim(-2, 1.7)
bx.set_ylim(-3, 3)
bx.set_xticks([-2, 0, 2])
bx.set_yticks([-2, 0, 2])
bx.set_title('Gráfica 2')

plt.show()
```

**9) (10 pts) Leer todos los datos de ventas de productos y mostrarlos usando un gráfico apilado (stack plot). El gráfico generado debe ser verse como el que se muestra de ejemplo.**

```python
#Escribe tu respuesta aquí
a = data.columns[1:7].tolist()

x = data['numero_mes']
y = data[a].values.T

labels =['Crema Facial', 'Jabón Facial', 'Pasta Dental', 'Jabón de Baño', 'Shampoo', 'Crema Hidratante']

plt.stackplot(x, y, 
              labels=labels,
              edgecolor='black')

plt.legend(loc=2)
plt.ylim(0, 39999)
plt.show()
```
