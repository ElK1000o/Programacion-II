```python

import numpy as np
#np.seterr(divide='ignore', invalid='ignore')
import matplotlib
import matplotlib.pyplot as plt


files2 = '../for007.dat'
for007F = np.genfromtxt(files2)
stavinf = 400.
nn = len(for007F[:,0])
grmax = max(for007F[:,3]);
glinemax = max(for007F[:,8]);

#Plots

plt.figure(0,figsize=(8.5,11.)) 
  #
left  = 0.03  # the left side of the subplots of the figure
right = 0.98    # the right side of the subplots of the figure
bottom = 0.05   # the bottom of the subplots of the figure
top = 0.97      # the top of the subplots of the figure
wspace = 0.01   # the amount of width reserved for blank space between subplots
hspace = 0.35   # the amount of height reserved for white space between subplot
plt.subplots_adjust(left=left, bottom=bottom, right=right, top=top, wspace=wspace, hspace=hspace)
  #
  #v(r) 
prange= [-1.01,0.0,0.0,400+50]
aspectrec=(prange[1]-prange[0])/(prange[3]-prange[2])*0.6
p1 = plt.subplot2grid((4,2), (0,0),aspect=aspectrec)
p1.plot(for007F[:,1],for007F[:,2],linestyle='-',color='b', label='HYDWIND');
p1.set_title('Velocity Profile',fontsize=8, fontweight='bold')
p1.set_xlabel('col 2',fontsize=7.5);
p1.set_ylabel('col 3',fontsize=7.5,labelpad=1) 
p1.tick_params(labelsize=7.5)
p1.axis(prange)
p1.legend(loc='upper left',frameon=False,fontsize=6.3,handlelength=2)
p1.grid(True,linestyle=':')  
  #
  #inner
prange= [0.999,1.04,0.0,20]
aspectrec=(prange[1]-prange[0])/(prange[3]-prange[2])*0.6
p2 = plt.subplot2grid((4,2), (0,1),aspect=aspectrec)
p2.plot(for007F[:,0],for007F[:,2],linestyle='-',color='b', label='HYDWIND');
p2.set_title('V(r/R$_{*}$) (inner)',fontsize=8, y=0.99, fontweight='bold')
p2.set_xlabel('col 1',fontsize=7.5);
p2.set_ylabel('col 3',fontsize=7.5,labelpad=-0);
p2.tick_params(labelsize=7.5)
p2.axis(prange)
p2.grid(True,linestyle=':')  
  #
  #outer
prange= [70.0,100,stavinf-20,stavinf+20]
aspectrec=(prange[1]-prange[0])/(prange[3]-prange[2])*0.6
p3 = plt.subplot2grid((4,2), (1,1),aspect=aspectrec)
p3.plot(for007F[:,0],for007F[:,2],linestyle='-',color='b', label='HYDWIND');
p3.set_title('V(r/R$_{*}$) (outer)',fontsize=8, y=0.99, fontweight='bold')
p3.set_xlabel('col 1',fontsize=7.5);
p3.set_ylabel('col 3',fontsize=7.5,labelpad=-0);
p3.tick_params(labelsize=7.5)
p3.axis(prange)
p3.grid(True,linestyle=':') 
  #
  #numeric and trial
rmax= max([stavinf,for007F[-1,16]])
prange= [-1.01,0.0,0.0,rmax+50]
aspectrec=(prange[1]-prange[0])/(prange[3]-prange[2])*0.6
p3 = plt.subplot2grid((4,2), (1,0),aspect=aspectrec)
p3.plot(for007F[:,1],for007F[:,2],linestyle='-',color='b', label='HYDWIND');
p3.plot(for007F[:,1],for007F[:,16],linestyle='-',color='r', label='TRIAL PROFILE');  
p3.set_title('Velocity Profile',fontsize=8, fontweight='bold')
p3.set_xlabel('col 2',fontsize=7.5);
p3.set_ylabel('col 3 (blue); col 17 (red)',fontsize=7.5,labelpad=1) 
p3.tick_params(labelsize=7.5)
p3.axis(prange)
p3.legend(loc='upper left',frameon=False,fontsize=6.3,handlelength=2)
p3.grid(True,linestyle=':')      
  #
  #dw/du
prange= [-1.01,0.0,0.0,grmax + 50.0]
aspectrec=(prange[1]-prange[0])/(prange[3]-prange[2])*0.6
p5 = plt.subplot2grid((4,2), (2,0),aspect=aspectrec)
p5.plot(for007F[:,1],for007F[:,3],linestyle='-',color='b', label='HYDWIND');
p5.set_title('dw/du',fontsize=8, fontweight='bold');
p5.set_xlabel('col 2',fontsize=7.5);
p5.set_ylabel('col 4',fontsize=7.5,labelpad=-0);
p5.tick_params(labelsize=7.5)
p5.axis(prange)
p5.grid(True,linestyle=':')  
  #
  #rho
prange= [-4, 2.1,np.log10(min(for007F[1:,5])*0.5),np.log10(max(for007F[1:,5])*10.)]
aspectrec=(prange[1]-prange[0])/(prange[3]-prange[2])*0.6
p7 = plt.subplot2grid((4,2), (2,1),aspect=aspectrec)
p7.plot(np.log10(for007F[:,0]-1),np.log10(for007F[:,5]),linestyle='-',color='b', label='HYDWIND');
p7.set_title('Density',fontsize=8, fontweight='bold');
p7.set_xlabel('log (col 1 - 1.0)',fontsize=7.5);
p7.set_ylabel('log (col 6)',fontsize=7.5,labelpad=-0);
p7.tick_params(labelsize=7.5)
p7.axis(prange)
p7.grid(True,linestyle=':')  
  #
# gline  
prange= [-1.0,20.0,-200, 1600]
aspectrec=(prange[1]-prange[0])/(prange[3]-prange[2])*0.6
p8 = plt.subplot2grid((4,2), (3,0),aspect=aspectrec)
p8.plot(for007F[:,0],for007F[:,8],linestyle='-',color='b', label='HYDWIND');
p8.set_title('g_line',fontsize=8, fontweight='bold');
p8.set_xlabel('col 1',fontsize=7.5);
p8.set_ylabel('col 9',fontsize=7.5,labelpad=-0);
p8.tick_params(labelsize=7.5)
p8.axis(prange)
p8.grid(True,linestyle=':')   

plt.savefig('solucion.pdf')

```
