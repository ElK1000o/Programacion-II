#### Importing Libraries and Data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

t_mov= pd.read_csv('../07-clase/datasets/types_movies.csv')
```

### Gráfico circular
El gráfico circular muestra la distribución de los datos en función de la proporción de la torta ocupada.

Es una gráfica circular que se divide en rebanadas para ilustrar la proporción numérica. La porción de un gráfico circular es para mostrar la proporción de partes de un todo.

Cuándo usar: el gráfico circular se debe usar con poca frecuencia, ya que es difícil comparar secciones del gráfico. En su lugar, se usa el gráfico de barras, ya que comparar secciones es fácil. ej.: Cuota de mercado en Películas.

```python
plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'])

plt.axis('equal')

plt.show()
```

Arriba, plt.pie() toma los datos numéricos como primer argumento, es decir, porcentaje y etiquetas para mostrar como segundo argumento, es decir, sector. En última instancia, muestra la distribución de datos en proporción al pastel.

Nota: los gráficos circulares no son tan buenos para ilustrar información. https://www.data-to-viz.com/caveat/pie.html 

### Componentes del gráfico circular
* <b>wedges:</b> Una lista de objetos Patch que representan cada cuña
* <b>texts:</b> Lista de objetos de texto que representan las etiquetas
* <b>autotexts:</b> Lista de objetos de texto para los valores numéricos: solo está disponible si el valor de autopct para el gráfico circular no es None

```python
wedges, texts, autotexts = plt.pie(t_mov['Percentage'], 
                                    labels=t_mov['Sector'],
                                    autopct='%.1f')

plt.axis('equal')

print('Wedges: ', wedges)
print('Texts: ', texts)
print('Autotexts: ', autotexts)
```

Desde arriba, podemos ver los componentes que hacen un gráfico circular y devuelve un objeto de cuña, texto en etiquetas, etc.

#### Personalizaciones de gráficos circulares
  Establecemos los valores para los colores y las propiedades de autopct. Este último establece el formato de visualización de los valores.

```python
colors = ['darkorange', 'sandybrown', 'darksalmon', 'orangered','chocolate']

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2i')

plt.axis('equal')

plt.show()
```

Un gráfico circular se puede personalizar fácilmente y desde arriba se formatean los valores de color y etiqueta.

#### La propiedad de explotar
Para resaltar una porción particular del gráfico circular, usamos explotar para separarla del resto del gráfico.

El valor de "explotar" representa la fracción del radio con el que compensar cada cuña.

```python
explode = [0, 0.0, 0, 0, 0.9]

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2f',
        explode=explode)

plt.axis('equal')

plt.show()
```
