# Stack Plot and Stem Plot
## Gráfico de pila

El ( Stack Plot) diagrama de pila visualiza los datos en pilas y muestra la distribución de los datos a lo largo del tiempo.

Cuándo usar: se usa para verificar múltiples regiones de área variable en una sola área.

Por ejemplo: es útil para comprender el cambio de distribución en múltiples variables durante un intervalo.

#### Importing Libraries and Data

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np_data= pd.read_csv('datasets/national_parks.csv')
```

```python
np_data.head()
```

Como el gráfico de pila requiere apilamiento, se hace usando np.vstack()

```python
x = np_data['Year']
```

```python
y = np.vstack([np_data['Badlands'], 
               np_data['GrandCanyon'], 
               np_data['BryceCanyon']])
```

### Stack Plots
Se utilizan diagramas de pila y visualización de datos a lo largo del tiempo.

```python
# Labels for each stack
labels = ['Badlands', 
          'GrandCanyon', 
          'BryceCanyon']

# Colors for each stack
colors = ['sandybrown', 
          'tomato', 
          'skyblue']

#  Similar to pandas df.plot.area()
plt.stackplot(x, y, 
              labels=labels,
              colors=colors, 
              edgecolor='black')

# Plots legend to the upperleft of Figure
plt.legend(loc=2)

plt.show()
```

plt.stackplot toma el primer argumento de datos numéricos, es decir, el año y el segundo argumento, los datos apilados verticalmente, es decir, los parques nacionales.

### Percentage Stacked Area Chart
Gráficos de pila basados en el porcentaje de distribución.

```python
plt.figure(figsize=(10,7))

data_perc = np_data.divide(np_data.sum(axis=1), axis=0)

plt.stackplot(x,
              data_perc["Badlands"],data_perc["GrandCanyon"],data_perc["BryceCanyon"],
              edgecolor='black',
              colors=colors,
              labels=labels)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

data_prec se utiliza para dividir el porcentaje general en distribuciones porcentuales individuales. s= np_data.sum(axis=1) calcula la suma a lo largo de las columnas, np_data.divide(s,axis=0) divide los datos a lo largo de las filas.

### Stem Plots
Similar a Stack Plots pero nos ayuda a ver números negativos, es decir, la diferencia en los datos a lo largo del tiempo.

Stemplot incluso toma valores negativos, por lo que la diferencia se toma de los datos y se grafica a lo largo del tiempo.

Cuándo usarlo: es similar a un diagrama de pila, pero la diferencia ayuda a comparar los puntos de datos.

```python
np_stem= np_data.copy()
```

```python
np_stem[['Badlands', 
         'GrandCanyon',
         'BryceCanyon']] = np_data[['Badlands', 
                                    'GrandCanyon',
                                    'BryceCanyon']].diff()

np_stem.head()
```

diff() se usa para encontrar la diferencia entre los datos anteriores y se almacena en otra copia de los datos. El primer punto de datos es NaN (No es un número), ya que no contiene ningún dato previo para calcular la diferencia.

```python
plt.figure(figsize=(15,11))

plt.suptitle('Change in Number of Visitors', y=0.94)

plt.subplot(311)
plt.stem(np_stem['Year'],
         np_stem['Badlands'],
         markerfmt = 'b_',
         linefmt = 'r--',
         basefmt = 'g:')
plt.title('GrandCanyon')

plt.subplot(312)
plt.stem(np_stem['Year'],
         np_stem['GrandCanyon'],
         markerfmt = 'b_',
         linefmt = 'r--',
         basefmt = 'g:')
plt.title('Badlands')

plt.subplot(313)
plt.stem(np_stem['Year'],
         np_stem['BryceCanyon'],
         markerfmt = 'b_',
         linefmt = 'r--',
         basefmt = 'g:')
plt.title('BryceCanyon')

plt.show()
```

(31n) Las subparcelas se crean para acomodar subparcelas de 3 filas y 1 columna en la figura. plt.stem() toma el primer argumento como datos numéricos, es decir, el año y el segundo argumento como datos numéricos de los visitantes del Parque Nacional.

# Gráfico de barra

Gráfico de barras muestra la distribución de datos en varios grupos. Comúnmente se confunde con un histograma que solo toma datos numéricos para trazar. Ayuda a comparar múltiples valores numéricos.

Cuándo usar: Se usa cuando se compara entre varios grupos.

Ej: Notas de los estudiantes en un examen.

#### Importing Libraries and Data

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

countries=pd.read_csv('datasets/countries.csv')
```

```python
countries.head()
```

```python
data_2007=countries[countries['year']==2007]
```

```python
top_10_p=data_2007.sort_values('population',ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))

x=range(10)
plt.bar(x,top_10_p['population']/10**9)
plt.xticks(x,top_10_p['country'])
plt.xlabel('Countries')
plt.ylabel('Population in Billions')
plt.title('10 Most Populous Countries in 2007')
plt.show()
```

plt.bar() toma el primer argumento como etiquetas en formato numérico y el segundo argumento para el valor que representa en las gráficas.

# Gráfico de dispersión

El diagrama de dispersión ayuda a visualizar 2 variables numéricas. Ayuda a identificar la relación de los datos con cada variable, es decir, patrones de correlación o tendencia. También ayuda a detectar valores atípicos en la trama.

Cuándo usarlo: se usa en conceptos de aprendizaje automático como la regresión, donde x e y son variables continuas. También se utiliza en agrupaciones de dispersión o detección de valores atípicos.

#### Importing Libraries and Data

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

countries=pd.read_csv('datasets/countries.csv')
```

```python
countries.head()
```

```python
data_2007=countries[countries['year']==2007]
```

# Scatter Plot

```python
plt.scatter(data_2007['gdpPerCapita'],data_2007['lifeExpectancy'])
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
data_2007['gdpPerCapita'].corr(data_2007['lifeExpectancy'])
```

plt.scatter() toma 2 argumentos numéricos para dispersar puntos de datos en el gráfico. Es similar al gráfico de líneas, excepto que no tiene las líneas rectas conectadas. Por corr nos referimos a correlación y significa que cuán correlacionado está el PIB con la esperanza de vida, como podemos ver que es positivo, significa que a medida que aumenta el PIB de un país, la esperanza de vida también aumenta.

```python
plt.scatter(np.log10(data_2007['gdpPerCapita']),data_2007['lifeExpectancy'])
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
np.log10(data_2007['gdpPerCapita']).corr(data_2007['lifeExpectancy'])
```

Al tomar el logaritmo del PIB, podemos que haya una correlación mucho mejor, ya que podemos ajustar mejor los puntos, convierte el PIB en escala logarítmica, es decir, log ($ 1000) = 3.

## Diagrama de dispersión 3D

El diagrama de dispersión 3D ayuda a visualizar 3 variables numéricas en un diagrama tridimensional.

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(countries['year'],countries['gdpPerCapita'],countries['lifeExpectancy'], s=30)
plt.show()
```

Es similar a la dispersión excepto que esta vez agregamos 3 variables numéricas. Al observar el gráfico, podemos inferir que a medida que aumenta el año y el PIB, también aumenta la esperanza de vida.
