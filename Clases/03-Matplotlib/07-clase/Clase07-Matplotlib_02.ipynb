# Histograma

Un histograma toma una serie de datos y los divide en varios contenedores. Luego traza los puntos de datos de frecuencia en cada contenedor (es decir, el intervalo de puntos). Es útil para comprender el conteo de rangos de datos.

Cuándo usar: Debemos usar el histograma cuando necesitamos el conteo de la variable en un gráfico. ej.: Número de juegos particulares vendidos en una tienda.

```python
plt.hist(np_data['GrandCanyon'], 
         facecolor='b', 
         edgecolor='k', 
         bins=10)
plt.show()
```

Desde arriba podemos ver el histograma de los visitantes del Gran Cañón en años. plt.hist() toma el primer argumento como datos numéricos en el eje horizontal, es decir, GrandCanyon visitor.bins=10 se usa para crear 10 contenedores entre los valores de los visitantes en GrandCanyon.

### Componentes de un histograma
* <b>n:</b> Contiene la frecuencia de cada bin.
* <b>bins:</b> Representa el valor medio de cada contenedor.
* <b>patches:</b> El objeto Patch para la forma de rectángulo que representa cada barra.

```python
n, bins, patches = plt.hist(np_data['GrandCanyon'], 
                            facecolor='peru', 
                            edgecolor='blue', 
                            bins=10)

print('n: ', n)
print('bins: ', bins)
print('patches: ', patches)
```

Desde arriba, podemos ver los componentes que forman un histograma, n como los valores máximos en cada intervalo del histograma, es decir, 5,9, y así sucesivamente.

#### La propiedad acumulativa
La propiedad acumulativa nos da el valor agregado final y nos ayuda a comprender el aumento de valor en cada contenedor.

Si es Verdadero, se calcula un histograma en el que cada contenedor proporciona los recuentos en ese contenedor más todos los contenedores para valores más pequeños. El último contenedor proporciona el número total de puntos de datos.

```python
plt.hist(np_data['GrandCanyon'], 
         facecolor='peru', 
         edgecolor='blue', 
         bins=10,
         cumulative=True)
plt.show()
```

#### Verifique el histograma a un rango de valores
Solo observamos los puntos de datos dentro del rango 2M-5M. Esto realinea los contenedores en el histograma.

El rango nos ayuda a comprender la distribución de valores entre valores especificados.

```python
plt.hist(np_data['GrandCanyon'], 
         facecolor='peru', 
         edgecolor='blue', 
         bins=10,
         range=(2000000, 3000000))

plt.show()
```

## Múltiples histogramas

Los histogramas múltiples son útiles para comprender la distribución entre 2 variables de entidad. Podemos ver que GrandCanyon tiene comparativamente más visitantes que BryceCanyon.

```python
plt.hist(np_data['BryceCanyon'], 
         facecolor='orangered', 
         edgecolor='maroon', 
         bins=8)

plt.hist(np_data['GrandCanyon'], 
         facecolor='peru', 
         edgecolor='blue', 
         bins=10, 
         alpha = 0.5)

plt.show()
```

## Experimente con un histograma 2d similar a los puntos de trazado, es decir, plt.plot (x, y)

Los histogramas 2D son útiles cuando necesita analizar la relación entre 2 variables numéricas que tienen una gran cantidad de valores. Es útil para evitar diagramas de dispersión demasiado trazados.

```python
# Big bins
plt.hist2d(np_data['Year'], np_data['GrandCanyon'], bins=(30, 30), cmap='jet')
plt.colorbar()
plt.show()
```

```python
# Big bins
plt.hist2d(np_data['Year'], np_data['Badlands'], bins=(10, 10), cmap='rainbow')
plt.colorbar()
plt.show()
```

```python
# Big bins
plt.hist2d(np_data['Year'], np_data['BryceCanyon'], bins=(50, 50), cmap=plt.cm.jet)
plt.colorbar()
plt.show()
```
