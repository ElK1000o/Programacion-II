# Serie de tiempo por diagrama de línea

La serie temporal es un gráfico de líneas y básicamente conecta puntos de datos con una línea recta. Es útil para comprender la tendencia a lo largo del tiempo. Puede explicar la correlación entre puntos por la tendencia. Una tendencia al alza significa una correlación positiva y una tendencia a la baja significa una correlación negativa. Se utiliza principalmente en pronósticos, modelos de monitoreo.

Cuándo usar: la serie temporal debe usarse cuando se van a trazar variables únicas o múltiples a lo largo del tiempo.

ej: Análisis Bursátil de Empresas, Pronóstico del Tiempo.

#### Importing Libraries and Data

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

stock_data = pd.read_csv('../07-clase/datasets/stocks.csv')
```

```python
stock_data.head()
```

Primero, convierta la fecha en pandas DateTime para facilitar el trazado de datos.

```python
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data.head()
```

#### Compare Stock side-by-side

```python
fig = plt.figure(figsize=(10,6))

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])

ax1.set_title('AAPL vs IBM(inset)')

ax1.plot(stock_data['Date'],
         stock_data['AAPL'],
         color='green')

ax2.plot(stock_data['Date'],
         stock_data['IBM'], 
         color='blue')
plt.show()
```

Desde arriba, fig.add_axes se usa para trazar el lienzo. Compruebe esto ¿Cuáles son las diferencias entre add_axes y add_subplot? comprender ejes y subtramas. plt.plot() toma el primer argumento como datos numéricos, es decir, Fecha y el segundo argumento es para datos numéricos de acciones. AAPL Stock se considera como ax1, que es la figura exterior y en ax2, IBM Stock se considera para trazar, que está insertado.

#### Compare multiple stocks

```python
# Figsize for width and height of plot
fig = plt.figure(figsize=(15,7))

fig.suptitle('Stock price comparison 2007-2017', 
             fontsize=20)

ax1 = fig.add_subplot(231)
ax1.set_title('MSFT')

ax1.plot(stock_data['Date'],
         stock_data['MSFT'], 
         color='green')

ax2 = fig.add_subplot(232)
ax2.set_title('GOOG')

ax2.plot(stock_data['Date'],
         stock_data['GOOG'], 
         color='purple')

ax3 = fig.add_subplot(233)
ax3.set_title('SBUX')

ax3.plot(stock_data['Date'],
         stock_data['SBUX'], 
         color='magenta')

ax4 = fig.add_subplot(234)
ax4.set_title('ADBE')

ax4.plot(stock_data['Date'],
         stock_data['ADBE'], 
         color='orange')

ax4 = fig.add_subplot(235)
ax4.set_title('NFLX')

ax4.plot(stock_data['Date'],
         stock_data['NFLX'], 
         color='chocolate')

ax4 = fig.add_subplot(236)
ax4.set_title('ORCL')

ax4.plot(stock_data['Date'],
         stock_data['ORCL'], 
         color='teal')

plt.show()
```

En la figura anterior, add_axes se usa para agregar ejes a una figura, mientras que arriba, add_subplot agrega varias subtramas a una figura. fig.add_subplot(237) no se puede hacer ya que solo hay 6 subtramas posibles.

Podemos ver que las acciones de las empresas tecnológicas siguen una tendencia alcista que muestra resultados positivos para que los comerciantes inviertan en acciones.

# Boxplot y Violinplot
## Diagrama de caja

Boxplot da un buen resumen de los datos. Ayuda a comprender mejor nuestra distribución.

Cuándo usar: Debe usarse cuando requerimos usar la información estadística general sobre la distribución de los datos. Se puede utilizar para detectar valores atípicos en los datos.

por ejemplo: puntuación de crédito del cliente. Podemos obtener el máximo, mínimo y mucha más información sobre la marca.

## Comprendiendo el diagrama de caja

![Image of Yaktocat](https://miro.medium.com/max/482/1*fCE_5juz235c6cmaOP_PDQ.png)

Del diagrama anterior, la línea que divide el cuadro en 2 partes representa la mediana de los datos. El final del cuadro muestra el cuartil superior (75 %) y el inicio del cuadro representa el cuartil inferior (25 %). El cuartil superior también se denomina 3er cuartil y, de manera similar, el cuartil inferior también se denomina 1er cuartil. La región entre el cuartil inferior y el cuartil superior se denomina Rango entre cuartiles (IQR) y se utiliza para aproximar la dispersión del 50 % en los datos intermedios (75–25=50 %). El máximo es el valor más alto en los datos, de manera similar, el mínimo es el valor más bajo en los datos, también se denomina límites. Los puntos fuera de las cajas y entre el máximo y el máximo se denominan bigotes y muestran el rango de valores en los datos. Los puntos extremos son valores atípicos de los datos. Una regla comúnmente utilizada es que un valor es un valor atípico si es inferior al cuartil inferior: 1,5 * IQR o superior al cuartil superior + 1,5 * IQR.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

exam_data = pd.read_csv('../07-clase/datasets/exams.csv')
```

```python
exam_data.head()
```

#### Extracting only the exam scores

```python
exam_scores = exam_data[['math score', 'reading score', 'writing score']]
exam_scores.head()
```

```python
exam_scores.describe()
```

```python
# For plotting in boxplot, we convert it to an array
exam_scores_array = np.array(exam_scores)
exam_scores_array
```

###  Box plots

```python
colors = ['blue', 'grey', 'lawngreen']

bp = plt.boxplot(exam_scores_array, 
                 patch_artist=True,
                 notch=True)

for i in range(len(bp['boxes'])):
    
    bp['boxes'][i].set(facecolor=colors[i])
    
    bp['caps'][2*i + 1].set(color=colors[i])
    
plt.xticks([1, 2, 3], ['Math', 'Reading', 'Writing'])

plt.show()
```

bp contiene los componentes del diagrama de caja como cajas, bigotes, medianas, tapas. Seaborn, otra biblioteca de gráficos, facilita la creación de gráficos personalizados que matplotlib. patch_artist hace posible la personalización. El notch que la mediana se vea más prominente.

Una advertencia de usar boxplot es que la cantidad de observaciones en el valor único no está definida, Jitter Plot en Seaborn puede superar esta advertencia o Violinplot también es útil

### Violin Plots

El diagrama de violín es un gráfico mejor que el diagrama de caja, ya que brinda una comprensión mucho más amplia de la distribución. Se asemeja a un violín y las áreas densas señalan la mayor distribución de datos que, de lo contrario, estarían ocultos por diagramas de caja.

Cuándo usar: Es una extensión de boxplot. Debe usarse cuando se requiere una mejor comprensión intuitiva de los datos.

```python
vp = plt.violinplot(exam_scores_array,
                    showmedians=True)

plt.xticks([1, 2, 3], ['Math', 'Reading', 'Writing'])

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor=colors[i])

plt.show()    
```

La densidad de puntos en el medio parece más, ya que los estudiantes tienden a obtener puntajes alrededor del promedio en su mayoría en las materias.

## TwinAxis

TwinAxis ayuda a visualizar el trazado de 2 gráficos en el eje y y el mismo eje x.

Cuándo usar: Debería cuando necesitamos 2  datos agrupados en la misma dirección.

Ej.: Población, datos del PIB en el mismo eje x (Fecha).

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

austin_weather = pd.read_csv('../07-clase/datasets/austin_weather.csv')
```

```python
austin_weather.head()
```

#### Extracting Date,Avg Temperature and Avg Wind Speed columns

```python
austin_weather = austin_weather[['Date', 'TempAvgF', 'WindAvgMPH']].head(30)
austin_weather
```

Extracción de detalles importantes, es decir, fecha para el eje x, TempAvgF y WindAvgMPH para los diferentes ejes y.

### Trazado de dos líneas con 2 ejes Y separados con el mismo eje X
Primero cree un eje y luego use la función twinx() para crear otro eje que comparta el mismo eje X

```python
 # Subplot for Plotting Figure
fig, ax_tempF = plt.subplots()  

# Similar to fig=plt.figure(figsize=(12,6))
fig.set_figwidth(12)
fig.set_figheight(6)

# set x label which is common
ax_tempF.set_xlabel('Date')

# bottom= false disables ticks and labelbottom disables x-axis labels
ax_tempF.tick_params(axis = 'x',
                    bottom=False,
                    labelbottom=False) 

# set left y-axis label
ax_tempF.set_ylabel('Temp (F)', 
                    color='red',
                    size='x-large')

# set labelcolor and labelsize to the left Y-axis
ax_tempF.tick_params(axis='y', 
                     labelcolor='red', 
                     labelsize='large')

# plot AvgTemp on Y-axis to the left 
ax_tempF.plot(austin_weather['Date'], 
              austin_weather['TempAvgF'], 
              color='red', label='TempAvgF')

#ax_tempF.legend(0)

# twinx sets the same x-axis for both plots 
ax_precip = ax_tempF.twinx()

#set Right y-axis label
ax_precip.set_ylabel('Avg Wind Speed (MPH)', 
                     color='blue', 
                     size='x-large') 

# set labelcolor and labelsize to the Right Y-axis
ax_precip.tick_params(axis='y', 
                      labelcolor='blue',
                      labelsize='large')

# plot WindAVg on Y-axis to the Right 
ax_precip.plot(austin_weather['Date'], 
         austin_weather['WindAvgMPH'], 
         color='blue', label= 'WindAvgMPH')

#ax_precip.legend()
#plt.legend()
fig.legend(loc=1, bbox_to_anchor=(1,1),bbox_transform=ax_tempF.transAxes)

plt.show()
```

Como podemos, solo hay 1 eje, twinx() se usa para hermanar el eje x y el eje y izquierdo se usa para Temp y el eje y derecho se usa para WindMPH.

### Eje gemelo para los mismos datos
Si los mismos datos se pueden expresar en diferentes unidades, podemos configurar nuestro eje para mostrar ambas unidades en dos ejes.

####  Función para convertir de fahrenheit a celsius

```python
def fahrenheit2celsius(f):
    return (f - 32)*5/9
```

```python
 # Subplot for Plotting Figure
fig, ax_tempF = plt.subplots()

# Similar to fig=plt.figure(figsize=(12,6))
fig.set_figwidth(12)
fig.set_figheight(6)

# set x label which is common
ax_tempF.set_xlabel('Date')

# bottom= false disables ticks and labelbottom disables x-axis labels
ax_tempF.tick_params(axis = 'x',
                    bottom=False,
                    labelbottom=False)

# set left y-axis label
ax_tempF.set_ylabel('Temp (F)', 
                    color='red',
                    size='x-large')

# set labelcolor and labelsize to the left Y-axis
ax_tempF.tick_params(axis='y', 
                     labelcolor='red', 
                     labelsize='large')

# plot AvgTemp on Y-axis to the left
ax_tempF.plot(austin_weather['Date'], 
              austin_weather['TempAvgF'], 
              color='green')

# twinx sets the same x-axis for both plots 
ax_tempC = ax_tempF.twinx()

# get ymin and ymin of AvgTemp in Fahrenheit
ymin, ymax = ax_tempF.get_ylim()

# set ymin and ymin of AvgTemp in Fahrenheit to AvgTemp in Celsius
ax_tempC.set_ylim((fahrenheit2celsius(ymin),
                   fahrenheit2celsius(ymax)))

#set Right y-axis label
ax_tempC.set_ylabel('Temp (C)', 
                    color='blue',
                    size='x-large')

# plot AvgTemp in Celsius on Y-axis to the Right 
ax_tempC.tick_params(axis='y', 
                     labelcolor='blue', 
                     labelsize='large')

plt.show()
```

Podemos ver que en el eje y de la izquierda se representa la temperatura en Fahrenheit y en el eje x de la derecha se representa la temperatura en Celsius.
